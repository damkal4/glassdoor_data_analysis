```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Damian\PycharmProjects\Praca dyplomowa\glassdoor_jobs.csv')
```

```python
df.head()
```

```python
df = df[df['Salary estimate'] != '-1']
```

```python
df['hourly'] = df['Salary estimate'].apply(lambda x: 1 if '/hr' in x.lower() else 0)
```

```python
salary = df['Salary estimate'].apply(lambda x: x.split('/')[0])
salary = salary.apply(lambda x: x.replace('$', ''))
df['salary'] = salary
```

```python
df['salary'] = df['salary'].apply(lambda x: x.replace(',', ''))
```

```python
df['salary'] = df['salary'].astype(float)
```

```python
df['salary'] = df.apply(lambda x: x.salary * 168 * 12 if x.hourly == 1 else x.salary, axis = 1)
```

```python
df['Founded'] = df['Founded'].apply(lambda x: -1 if x == 'Government' or x == 'Company - Public' 
                                    or x == 'Company - Private'
                                    or x == 'Subsidiary or Business Segment' else x )
```

```python
df['age'] = df.Founded.apply(lambda x: int(x) if int(x) < 1 else 2022 - int(x))
```

```python
df['Python'] = df['Job description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
df['SQL'] = df['Job description'].apply(lambda x: 1 if 'sql' in x.lower() else 0)
df['Pandas'] = df['Job description'].apply(lambda x: 1 if 'pandas' in x.lower() else 0)
df['R Studio'] = df['Job description'].apply(lambda x: 1 if 'r studio' in x.lower()
                                             or 'r-studio' in x.lower() else 0)
df['Excel'] = df['Job description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
df['AWS'] = df['Job description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
```

```python
df.to_csv('cleaned_data.csv')
```
